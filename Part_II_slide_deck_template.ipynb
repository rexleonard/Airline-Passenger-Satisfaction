## Investigation Overview


> In this investigation, I looked at the overall distribution of passenger satisfaction level to determine how well the passengers were satisfied with the airline. Also, the distribution of the passengers travel class.


## Dataset Overview

> The data set is an airline satisfaction scores for 129,880 passengers, were each record represents one passenger. The records contain details about passenger demographics, flight distance, delays, travel class, purpose, ratings for factors such as cleanliness, comfort, service, and as well as overall satisfaction with the airline.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
airline = pd.read_csv('airline_cleaned.csv')
```

> Note that the above cells have been set as "Skip"-type slides. That means

## (Distribution of Passengers Satisfaction)

> The passenger's satisfaction is the overall satisfaction level with the airline, a passenger can either be Satisfied/Neutral or unsatisfied. The distribution below displays the overall satisfaction level of passengers.


```python
# Plot showing Passenger Satisfaction
default_color = sb.color_palette()[1]
sb.countplot(data = airline, x = 'Satisfaction', color = default_color)
plt.xlabel('Satisfaction')
plt.title('Passenger Satisfaction')
```

## (Distribution of Passengers Class)

> There are three travel class in the dataset. The chart below shows the distribution of the travel class.

```python
# Distribution of the passenger class
default_color = sb.color_palette()[0]
sb.countplot(data = airline, x = 'Class', color = default_color)
plt.xlabel('Class')
plt.title('Passenger Class')
```

## (Departure Delay Vs Arrival Delay)

> From the plot of departure delay vs arrival delay, we can see the direct correlation between flight departure delay and its arrival delay time. 



```python
# scatter distribution of Departure Delay vs. Arrival Delay

plt.figure(figsize = [8, 6])
plt.scatter(data = airline, x = 'Departure Delay', y = 'Arrival Delay', alpha = 1/5)
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.title('Scatter plot of Departure Delay vs. Arrival Delay')
plt.show()
```
